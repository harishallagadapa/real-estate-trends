```python
def convert_price(price):
    if isinstance(price, float):
        return price
    price = price.replace(',', '').replace('₹', '')
    if 'Lac' in price:
        price = price.replace('Lac', '')
        return float(price) * 100000
    elif 'Cr' in price:
        price = price.replace('Cr', '')
        return float(price) * 10000000
    else:
        return float(price.strip())

df2['Price(in ₹)'] = df2['Price(in ₹)'].apply(convert_price)
```

```python
import pymysql

# Establish a database connection
connection = pymysql.connect(
    host='127.0.0.1',
    port=3306,
    user='root',
    password='password',
    db='Real_Estate_Market'
)

# Create a cursor
cursor = connection.cursor()

# Define your CREATE TABLE query
create_table_query = """
CREATE TABLE IF NOT EXISTS Houses (
    id INT AUTO_INCREMENT PRIMARY KEY,
    HouseTitle VARCHAR(255),
    Builder VARCHAR(255),
    Location VARCHAR(255),  # Make sure this matches your dataframe column name
    Furnishing VARCHAR(255),
    Bathroom INT,
    TenantPrefered VARCHAR(255),
    Availability VARCHAR(255),
    Area VARCHAR(255),
    Floor VARCHAR(255),
    PropertyFacing VARCHAR(255),
    Notes TEXT,
    Rent VARCHAR(255),
    NearBy VARCHAR(255),
    Link VARCHAR(255)
)
"""

# Execute the CREATE TABLE query
cursor.execute(create_table_query)

# Iterate over your dataframe and insert data
for _, row in df.iterrows():
    insert_query = """
    INSERT INTO Houses (HouseTitle, Builder, Location, Furnishing, Bathroom, TenantPrefered, Availability, Area, Floor, PropertyFacing, Notes, Rent, NearBy, Link)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    cursor.execute(insert_query, (
        row['HouseTitle'],
        row['Builder'],
        row['Location'],  
        row['Furnishing'],
        row['Bathroom'],
        row['TenantPrefered'],
        row['Availability'],
        row['Area'],
        row['Floor'],
        row['PropertyFacing'],
        row['Notes'],
        row['Rent'],
        row['NearBy'],
        row['Link']
    ))

# Commit changes
connection.commit()
```
